```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  Load the CSV file
df = pd.read_csv("Pakistan_Poverty_Dataset_2000_2023 (1).csv")
```

```python
# Clean the data
df.dropna(inplace=True)
```

```python
# Rename columns for easier access
df.rename(columns={
    'Unemployment Rate (%)': 'Unemployment',
    'Year': 'Year'
}, inplace=True)
```

```python
# Plot: Unemployment Rate Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Unemployment', marker='o', color='blue')
plt.title(' Unemployment Rate in Pakistan (20002023)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# COVID-19 Period: 20192021
covid_years = df[df['Year'].between(2019, 2021)]
```

```python
plt.figure(figsize=(8, 7))
sns.barplot(data=covid_years, x='Year', y='Unemployment', hue='Year', palette='Reds', legend=False)
plt.title(' Unemployment Spike During COVID-19 (20192021)', fontsize=13)
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title(' Correlation Between Economic Indicators', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set Seaborn style
sns.set(style="whitegrid")
```

```python
# Load dataset
df = pd.read_csv("Pakistan_Poverty_Dataset_2000_2023 (1).csv")
df.dropna(inplace=True)
df.rename(columns={'Unemployment Rate (%)': 'Unemployment'}, inplace=True)
```

```python
# Clean the data
df.dropna(inplace=True)
```

```python
# --- 1. Line Plot: Unemployment Over Time ---
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Unemployment', data=df, marker='o', color='blue')
plt.title('Unemployment Rate in Pakistan (20002023)')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.tight_layout()
plt.savefig("unemployment_trend.png")  # Save plot as PNG
plt.show()
```

```python

# --- 2. Bar Chart: COVID Impact ---
covid_years = df[df['Year'].between(2019, 2021)]
plt.figure(figsize=(8, 5))
sns.barplot(data=covid_years, x='Year', y='Unemployment', hue='Year', palette='Reds', legend=False)
plt.title('COVID-19 Impact on Unemployment')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.tight_layout()
plt.savefig("covid_unemployment.png")
plt.show()
```

```python

# --- 3. Multi-Line Plot: Unemployment vs Inflation vs Poverty ---
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Unemployment', data=df, label='Unemployment')
sns.lineplot(x='Year', y='Inflation Rate (%)', data=df, label='Inflation')
sns.lineplot(x='Year', y='Poverty Headcount Ratio (%)', data=df, label='Poverty')
plt.title('Unemployment vs Inflation vs Poverty (20002023)')
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("multi_indicator_comparison.png")
plt.show()
```

```python
# --- 4. Correlation Heatmap ---
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Economic Indicators')
plt.tight_layout()
plt.savefig("correlation_heatmap.png")
plt
```
